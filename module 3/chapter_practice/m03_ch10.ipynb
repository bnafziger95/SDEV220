```python
def get_odds():
    for number in range(1, 10):
        if number % 2 ==1:
            yield number

count = 1
for number in get_odds():
    if count == 3:
        print("The third odd number is: ", number)
        break
    count += 1
```

```python
def test(func):
    def new_func(*args, **kwargs):
        print('start')
        result = func(*args, **kwargs)
        print('end')
        return result
    return new_func

@test
def greeting():
    print("Greetings, Earthling")

greeting()
```
